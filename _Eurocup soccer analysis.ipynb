```python
grouped_teams = eurocup.groupby('Team')
```

```python
goals= grouped_teams['Goals'].sum()
goals.sort_values(ascending=False)
```

```python
def plot_goals(goals_data):
    # Sort the goals data in descending order
    sorted_data = goals_data.sort_values(ascending=False)

    # Set the style of the plot
    sns.set(style='whitegrid')

    # Create the bar plot using Seaborn
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sorted_data.values, y=sorted_data.index)

    # Set the title and labels
    plt.title('Goals Scored by Teams')
    plt.xlabel('Number of Goals')
    plt.ylabel('Team')

    # Display the plot
    plt.show()
```

```python
goals = grouped_teams['Goals'].sum()
plot_goals(goals)
```
